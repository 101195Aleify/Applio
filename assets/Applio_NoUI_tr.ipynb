```python
# @title Kurulum
rot_47 = lambda kodlanmış_metin: "".join(
    [
        (
            chr(
                (ord(h) - (ord("a") if h.islower() else ord("A")) - 47) % 26
                + (ord("a") if h.islower() else ord("A"))
            )
            if h.isalpha()
            else h
        )
        for h in kodlanmış_metin
    ]
)
import codecs
import os
import shutil
import tarfile
import subprocess
from pathlib import Path
from datetime import datetime
E = Exception
B = print


def vidal_kurulumu(Zorla):
    L = "Kikpm.ovm.bu"
    K = "/content/"
    C = Zorla

    def F():
        print("Pip paketleri yükleniyor...")
        subprocess.check_call(["pip", "install", "-r", "requirements.txt", "--quiet"])

    A = K + rot_47(L)
    G = K + rot_47(L)
    D = "/"
    if not os.path.exists(A):
        M = os.path.dirname(A)
        os.makedirs(M, exist_ok=True)
        print("Önbelleğe alınmış kurulum bulunamadı..")
        try:
            N = rot_47(
                codecs.decode(
                    "pbbxa://pcooqvonikm.kw/QIPqaxivw/Ixxtqw/zmawtdm/uiqv/Kwtij/Xvxcz.biz.oh",
                    "rot_13",
                )
            )
            subprocess.run(["wget", "-O", A, N])
            print("İndirme başarıyla tamamlandı!")
        except E as H:
            print(str(H))
            if os.path.exists(A):
                os.remove(A)
    if Path(A).exists():
        with tarfile.open(G, "r:gz") as I:
            for J in I.getmembers():
                O = os.path.join(D, J.name)
                try:
                    I.extract(J, D)
                except E as H:
                    print("Bir dosya çıkarılamadı")
                    C = True
            print(f"{G} dosyasının {D} konumuna çıkarılması tamamlandı.")
        if os.path.exists(A):
            os.remove(A)
        if C:
            F()
            C = False
    else:
        F()


vidal_kurulumu(False)
print("Gereksinimlerin kurulumu tamamlandı!")
```

```python
# @title Eğit
# @markdown ### ➡️ Model Bilgisi
import requests
import threading
import time
import os
import shutil
import hashlib
import time

LOGS_KLASÖRÜ = "/content/Applio/logs/"
AĞIRLIKLAR_KLASÖRÜ = LOGS_KLASÖRÜ + model_adı
GOOGLE_DRIVE_YOLU = "/content/drive/MyDrive/RVC_Backup"


def google_drive_yedeğini_al():
    print("Google Drive yedeği alınıyor...")
    ağırlıklar_mevcut = False
    for kök, dizinler, dosyalar in os.walk(GOOGLE_DRIVE_YOLU):
        for dosya_adı in dosyalar:
            dosya_yolu = os.path.join(kök, dosya_adı)
            if os.path.isfile(dosya_yolu) and not dosya_yolu.startswith(
                os.path.join(GOOGLE_DRIVE_YOLU, "ağırlıklar")
            ):
                yedek_dosya_yolu = os.path.join(
                    LOGS_KLASÖRÜ, os.path.relpath(dosya_yolu, GOOGLE_DRIVE_YOLU)
                )
                yedek_klasör_yolu = os.path.dirname(yedek_dosya_yolu)
                if not os.path.exists(yedek_klasör_yolu):
                    os.makedirs(yedek_klasör_yolu)
                    print(f"Yedek klasör oluşturuldu: {yedek_klasör_yolu}", flush=True)
                shutil.copy2(dosya_yolu, yedek_dosya_yolu)  # dosyayı meta verileriyle birlikte kopyala
                print(f"Google Drive yedeğinden dosya alındı: {dosya_adı}")
            elif dosya_yolu.startswith(
                os.path.join(GOOGLE_DRIVE_YOLU, "ağırlıklar")
            ) and dosya_adı.endswith(".pth"):
                ağırlıklar_mevcut = True
                ağırlıklar_dosya_yolu = os.path.join(
                    AĞIRLIKLAR_KLASÖRÜ,
                    os.path.relpath(
                        dosya_yolu, os.path.join(GOOGLE_DRIVE_YOLU, "ağırlıklar")
                    ),
                )
                ağırlıklar_klasör_yolu = os.path.dirname(ağırlıklar_dosya_yolu)
                if not os.path.exists(ağırlıklar_klasör_yolu):
                    os.makedirs(ağırlıklar_klasör_yolu)
                    print(
                        f"Ağırlıklar klasörü oluşturuldu: {ağırlıklar_klasör_yolu}",
                        flush=True,
                    )
                shutil.copy2(dosya_yolu, ağırlıklar_dosya_yolu)  # dosyayı meta verileriyle birlikte kopyala
                print(f"Ağırlıklardan dosya alındı: {dosya_adı}")
    if ağırlıklar_mevcut:
        print("Google Drive yedeğinden ağırlıklar yerel ağırlıklar klasörüne kopyalandı.")
    else:
        print("Google Drive yedeğinde ağırlıklar bulunamadı.")
    print("Google Drive yedek alma tamamlandı.")


def md5_hash_al(dosya_yolu):
    hash_md5 = hashlib.md5()
    with open(dosya_yolu, "rb") as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()


def ağırlıkları_drive_a_kopyala():
    hedef_klasör = os.path.join(GOOGLE_DRIVE_YOLU, "ağırlıklar")
    try:
        if not os.path.exists(hedef_klasör):
            os.makedirs(hedef_klasör)

        kopyalanan_sayısı = 0
        for dosya_adı in os.listdir(AĞIRLIKLAR_KLASÖRÜ):
            if dosya_adı.endswith(".pth"):
                kaynak_dosya = os.path.join(AĞIRLIKLAR_KLASÖRÜ, dosya_adı)
                hedef_dosya = os.path.join(hedef_klasör, dosya_adı)
                if not os.path.exists(hedef_dosya):
                    shutil.copy2(kaynak_dosya, hedef_dosya)
                    kopyalanan_sayısı += 1
                    print(f"{dosya_adı} Google Drive'a kopyalandı!")

        if kopyalanan_sayısı == 0:
            print("Kopyalanacak yeni tamamlanmış modeller bulunamadı.")
        else:
            print(f"{kopyalanan_sayısı} dosya Google Drive'a kopyalandı!")

    except Exception as e:
        print(f"Ağırlıklar kopyalanırken bir hata oluştu: {str(e)}")


if "otomatik_yedekler" not in globals():
    otomatik_yedekler = False


def dosyaları_yedekle():
    print("\nYedekleme döngüsü başlıyor...")
    son_yedek_zamanları_yolu = os.path.join(
        LOGS_KLASÖRÜ, "last_backup_timestamps.txt"
    )
    tam_olarak_güncellendi = False

    while True:
        try:
            güncellendi = False
            son_yedek_zamanları = {}

            try:
                with open(son_yedek_zamanları_yolu, "r") as f:
                    son_yedek_zamanları = dict(
                        line.strip().split(":") for line in f
                    )
            except FileNotFoundError:
                pass

            for kök, dizinler, dosyalar in os.walk(LOGS_KLASÖRÜ):
                # "zips" klasörünü hariç tut
                if "zips" in dizinler:
                    dizinler.remove("zips")
                if "mute" in dizinler:
                    dizinler.remove("mute")
                for dosya_adı in dosyalar:
                    if dosya_adı != "last_backup_timestamps.txt":
                        dosya_yolu = os.path.join(kök, dosya_adı)
                        if os.path.isfile(dosya_yolu):
                            yedek_dosya_yolu = os.path.join(
                                GOOGLE_DRIVE_YOLU,
                                os.path.relpath(dosya_yolu, LOGS_KLASÖRÜ),
                            )
                            yedek_klasör_yolu = os.path.dirname(yedek_dosya_yolu)
                            if not os.path.exists(yedek_klasör_yolu):
                                os.makedirs(yedek_klasör_yolu)
                                print(
                                    f"Yedek klasörü oluşturuldu: {yedek_klasör_yolu}",
                                    flush=True,
                                )
                            son_yedek_zamanı = son_yedek_zamanları.get(
                                dosya_yolu)
                            mevcut_zaman = os.path.getmtime(dosya_yolu)
                            if (
                                son_yedek_zamanı is None
                                or float(son_yedek_zamanı) < mevcut_zaman
                            ):
                                shutil.copy2(dosya_yolu, yedek_dosya_yolu)
                                son_yedek_zamanları[dosya_yolu] = str(
                                    mevcut_zaman
                                )
                                if son_yedek_zamanı is None:
                                    print(f"Dosya yedeklendi: {dosya_adı}")
                                else:
                                    print(
                                        f"Yedeklenen dosya güncellendi: {dosya_adı}")
                                güncellendi = True
                                tam_olarak_güncellendi = False

            for dosya_yolu in list(son_yedek_zamanları.keys()):
                if not os.path.exists(dosya_yolu):
                    yedek_dosya_yolu = os.path.join(
                        GOOGLE_DRIVE_YOLU, os.path.relpath(
                            dosya_yolu, LOGS_KLASÖRÜ)
                    )
                    if os.path.exists(yedek_dosya_yolu):
                        os.remove(yedek_dosya_yolu)
                        print(f"Silinen dosya: {dosya_yolu}")
                    del son_yedek_zamanları[dosya_yolu]
                    güncellendi = True
                    tam_olarak_güncellendi = False

            if not güncellendi and not tam_olarak_güncellendi:
                print("Dosyalar güncel.")
                tam_olarak_güncellendi = True
                uyku_süresi = 15
            else:
                uyku_süresi = 0.1

            with open(son_yedek_zamanları_yolu, "w") as f:
                for dosya_yolu, zaman_damgası in son_yedek_zamanları.items():
                    f.write(f"{dosya_yolu}:{zaman_damgası}\n")

            time.sleep(uyku_süresi)

        except Exception as e:
            print(f"Bir hata oluştu: {str(e)}")


if otomatik_yedekler:
    otomatik_yedekler = False
    print("Otomatik yedekleme devre dışı bırakıldı")
else:
    otomatik_yedekler = True
    print("Otomatik yedekleme etkinleştirildi")

toplam_çevrim = 800  # @param {type:"integer"}
yığın_boyutu = 15  # @param {type:"slider", min:1, max:25, step:0}
gpu = 0
örnek_oranı = int(örnek_oranı.rstrip("k")) * 1000
pitch_rehberliği = True  # @param{type:"boolean"}
önceden_eklenmiş = True  # @param{type:"boolean"}
tensorboard = True  # @param{type:"boolean"}
# @markdown ### ➡️ Modelinizin hangi döngüde saklanacağını seçin
her_çevrimde_sakla = 10  # @param {type:"slider", min:1, max:100, step:0}
yalnızca_sonuncuyu_sakla = False  # @param{type:"boolean"}
her_ağırlıkta_sakla = False  # @param{type:"boolean"}
aşırı_eğitim_izleyicisi = False  # @param{type:"boolean"}
aşırı_eğitim_eşiği = 50  # @param {type:"slider", min:1, max:100, step:0}
# @markdown ### ❓ Opsiyonel
# @markdown Özel önceden eğitilmiş seçerseniz, önceden eğitilmişleri indirmeniz ve önceden eğitilmişlerin yolunu girmeniz gerekir.
özel_önceden_eğitilmiş = False  # @param{type:"boolean"}
g_önceden_eğitilmiş_yolu = "/content/Applio/rvc/pretraineds/pretraineds_custom/G48k.pth"  # @param {type:"string"}
d_önceden_eğitilmiş_yolu = "/content/Applio/rvc/pretraineds/pretraineds_custom/D48k.pth"  # @param {type:"string"}

if "pretrained" not in globals():
    önceden_eklenmiş = True

if "custom_pretrained" not in globals():
    özel_önceden_eğitilmiş = False

if "g_pretrained_path" not in globals():
    g_önceden_eğitilmiş_yolu = "Custom Path"

if "d_pretrained_path" not in globals():
    d_önceden_eğitilmiş_yolu = "Custom Path"


def eğitimi_başlat():
    if tensorboard == True:
        %load_ext tensorboard
        %tensorboard --logdir /content/Applio/logs/
    !python core.py train --model_name "{model_adı}" --rvc_version "{rvc_sürümü}" --save_every_epoch "{her_çevrimde_sakla}" --save_only_latest "{yalnızca_sonuncuyu_sakla}" --save_every_weights "{her_ağırlıkta_sakla}" --total_epoch "{toplam_çevrim}" --sampling_rate "{örnek_oranı}" --batch_size "{yığın_boyutu}" --gpu "{gpu}" --pitch_guidance "{pitch_rehberliği}" --pretrained "{önceden_eklenmiş}" --custom_pretrained "{özel_önceden_eğitilmiş}" --g_pretrained_path "{g_önceden_eğitilmiş_yolu}" --d_pretrained_path "{d_önceden_eğitilmiş_yolu}" --overtraining_detector "{aşırı_eğitim_izleyicisi}" --overtraining_threshold "{aşırı_eğitim_eşiği}"


sunucu_işi = threading.Thread(target=eğitimi_başlat)
sunucu_işi.start()

if otomatik_yedekler:
    dosyaları_yedekle()
else:
    while True:
        time.sleep(10)
```
